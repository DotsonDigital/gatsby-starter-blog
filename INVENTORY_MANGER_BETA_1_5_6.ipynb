```python
# ################################################
#    ___________ ____  __ __ ____
#   / ____/ ___// __ \/ //_//  _/
#  / /    \__ \/ / / / ,<   / /
# / /___ ___/ / /_/ / /| |_/ /
# \____//____/\____/_/ |_/___/

#   ___                      _
#  |_ _| _ _ __ __ ___  _ _ | |_  ___  _ _  _  _
#   | | | ' \\ V // -_)| ' \|  _|/ _ \| '_|| || |
#  |___||_||_|\_/ \___||_||_|\__|\___/|_|   \_, |
#                                           |__/
#   __  __
#  |  \/  | __ _  _ _   __ _  __ _  ___  _ _
#  | |\/| |/ _` || ' \ / _` |/ _` |/ -_)| '_|
#  |_|  |_|\__,_||_||_|\__,_|\__, |\___||_|
#                            |___/
# ################################################
# ########## Contact jimmy@dotson.tech ###########
# ################################################
#
# Type in the actual inventory count of the item
# and it sends and email to jd@csoki with
# anything below the pre-defined threshold.
#
# ################################################
# ################################################
## Revisions:
## 1.5.5 Removed forwarding rules. Against CSOKIpolicy?
## -- SOLUTION: Manually copy and paste email to NGSupply.
##
## 1.5.6 Added num_papert (paper towels)
##



def request_supplies(
    num_paper, num_tpaper, num_papert, num_glasscleaner, num_spraycleaner,
    num_safepaper, num_cleaningwipes, num_trashbags, num_shoppingbags,
    num_staples, num_rubberbands, num_pens, num_postitnotes,
    num_handsanitizer, num_handsoap):
  """
  Creates an email requesting more supplies if any of the supplies are
  below the threshold.

  Args:
    num_paper: Number of packs of printer paper.
    num_tpaper: Number of toilet paper rolls in stock.

  Returns:
    A string containing the email to be sent.
  """

  email = "\n"

# What to write in the email, ONLY if below threshold
#=====================================================

  if num_paper <= threshold_paper:
    email += f"Printer Paper, {num_paper} in stock \n"

  if num_tpaper <= threshold_tpaper:
    email += f"Toilet paper, {num_tpaper} in stock \n"

  if num_papert <= threshold_paper:
    email += f"Paper Towels, {num_papert} in stock \n"

  if num_glasscleaner <= threshold_glasscleaner:
    email += f"Glass Cleaner, {num_glasscleaner} in stock \n"

  if num_spraycleaner <= threshold_spraycleaner:
    email += f"Spray Cleaner, {num_spraycleaner} in stock \n"

  if num_safepaper <= threshold_safepaper:
    email += f"Paper for the SAFE, {num_safepaper} in stock  \n"

  if num_cleaningwipes <= threshold_cleaningwipes:
    email += f"Cleaning Wipes, {num_cleaningwipes} in stock  \n"

  if num_trashbags <= threshold_trashbags:
    email += f"Trash Bags \n"

  if num_shoppingbags <= threshold_shoppingbags:
    email += f"Shopping Bags, {num_shoppingbags} in stock (Estimated)  \n"

  if num_staples < threshold_staples:
    email += f"Staples \n"

  if num_rubberbands <= threshold_rubberbands:
    email += f"Rubber Bands\n"

  if num_pens <= threshold_pens:
    email += f"Ink Pens \n"

  if num_postitnotes < threshold_postitnotes:
    email += f"Post-It Notes \n"

  if num_handsanitizer < threshold_handsanitizer:
    email += f"Hand Sanitizer, only {num_handsanitizer} found in {location} \n"

  if num_handsoap < threshold_handsoap:
    email += f"Hand Soap, only {num_handsoap} found in {location}. \n"

#Create the email content from above
#=====================================================

  if email:
    email = f"""

    Dear NGSupply,

    {location} needs the following items:
      {email}
    Please send more supplies to {location}. Thanks!


    Jimmy Dotson
    james.dotson@cellularsales.com


    """
    return email
  else:
    return None


# Get user count via input for each item
#=====================================================
location = input("What store are you at?: ")
num_paper = int(input("How many FULL packs of paper are there?: "))
threshold_paper = 4
num_tpaper = int(input("How many UNUSED rolls of toilet paper are there?: "))
threshold_tpaper = 5
num_papert = int(input("How many UNUSED rolls of paper towels can you find?: "))
threshold_papert = 4
num_glasscleaner  = int(input("How many FULL bottles of glass cleaner are there?: "))
threshold_glasscleaner = 2
num_spraycleaner = int(input("How many bottles of Spray Cleaner are there?: "))
threshold_spraycleaner = 1
num_safepaper = int(input("How many rolls of Safe Paper do you see?: "))
threshold_safepaper = 1
num_cleaningwipes = int(input("How many Cleaning Wipes in stock?: "))
threshold_cleaningwipes = 2
num_trashbags = int(input("How many FULL trash bags rolls can you count?: "))
threshold_trashbags = 2
num_shoppingbags = int(input("How many shopping bags can you estimate we have?: "))
threshold_shoppingbags = 100
num_staples = int(input("How many FULL boxes of Staples?: "))
threshold_staples = 1
num_rubberbands = int(input("How many Ruber bands can you estimate?: "))
threshold_rubberbands = 50
num_pens = int(input("How many blue ink pens are available?: "))
threshold_pens = 10
num_postitnotes = int(input("How many packs of Post-It notes can you find?: "))
threshold_postitnotes = 5
num_handsanitizer = int(input("How many FULL bottles of Hand Sanitizer do you see?: "))
threshold_handsanitizer = 4
num_handsoap = int(input("How many bottles of FULL handsoap can you find?: "))
threshold_handsoap = 2

# Create and send the email
#=====================================================

email = request_supplies(
    num_paper, num_tpaper, num_papert, num_glasscleaner, num_spraycleaner,
    num_safepaper, num_cleaningwipes, num_trashbags, num_shoppingbags,
    num_staples, num_rubberbands, num_pens, num_postitnotes,
    num_handsanitizer, num_handsoap)

## Show the result
#=====================================================

print(email)

######################################################
######  Uncomment below to fire up the emails   ######
######################################################

# Import the necessary libraries for email
#=====================================================
import smtplib

# Recipient email address(s)
#=====================================================
recipient_email = "james.dotson@cellularsales.com"

# Create the email message from the count
#=====================================================
message = f"Subject: Supply Request for {location}\n\n{email}"

# Send the damn email
#=====================================================
with smtplib.SMTP("smtp.gmail.com", 587) as smtp:
    smtp.starttls()
    smtp.login("jimmy.cellularsales@gmail.com", "lrkb jlcc rdhu vwyy")
    smtp.sendmail("jimmy.cellularsales@gmail.com", recipient_email, message)

# Print a confirmation message
#=====================================================
print("Email sent successfully to: ", recipient_email)
```
